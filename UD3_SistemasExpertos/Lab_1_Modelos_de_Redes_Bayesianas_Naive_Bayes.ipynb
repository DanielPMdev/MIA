**0. Librerias necesarias para el laboratorio.**

```python
import pandas as pd
import numpy as np


%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

**1. Análisis descriptivo de los datos:**

```python
df = pd.read_csv(r'lab1-dataset.txt')
# X Caracteristicas -> demás columnas
# Y Variable Objetivo -> classname

df.head()
```

```python
print(df.groupby('classname').size())
```

**2. Tratamiento de missing:**

```python
#df.describe()

for col in df.columns:
    df.fillna({col: df[col].mode()[0]}, inplace=True)
```

**3. Determinar el conjunto de modelización y el de validación.**

Buscamos dividir nuestros datos para que el 75 % de los datos sean para entrenamiento y el 25 % para test.


```python
# Dividir los datos en conjunto de entrenamiento y de test
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(df, random_state=2)

# Vemos que se ha realizado correctamente el reparto de los datos.
print("Ejemplos usados para entrenar: ", len(X_train))
print("Ejemplos usados para test: ", len(X_test))

# Asignaremos los valores de classname
y_train = X_train['classname']
y_test = X_test['classname']
print("Distribución de los datos de test:" )
print(X_test['classname'].value_counts(sort=True))
```

Nos ponemos a entrenar nuestros datos, para ello vamos a utilizar Multinomial Naive Bayes al ser datos discretos.

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()

X_train.drop(['classname'], axis=1)
naive_bayes.fit(X_train, y_train)
```

Por último vamos a predecir utilizando los datos test.

```python
X_test.drop(['classname'], axis=1)
predictions = naive_bayes.predict(X_test)
```

**4. Calcular las métricas de evaluación de ajuste adecuadas.**

```python
reportOriginal = precision_score(y_test, predictions)
print(reportOriginal)
```
